```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gradientDescent(pred, goal, weight, input):
    delta = pred - goal
    print(pred, "-", goal, "=", delta)
    weightDelta = input * delta * alpha
    print(input, "*", delta, "*", alpha, "=", weightDelta)
    print(weight, "-", weightDelta, "=", weight-weightDelta)
    weight -= weightDelta
    return weight

def neuralNetwork(input, weight):
    pred = input * weight
    return pred
```

```python
weight = 0.1
alpha = 0.01

games = [0]
winOrLose = [1]
input = games[0]
goal = winOrLose[0]

errs = []
preds = []
```

```python
for i in range(20):
    pred = neuralNetwork(input, weight)
    error = (pred - goal) ** 2
    
    preds.append(pred)
    errs.append(error)
    
    weight = gradientDescent(pred, goal, weight, input)
    
```
